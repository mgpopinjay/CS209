#### Problem 1(c)

Create a dictionary of pandas DataFrames called `polls` keyed by the name of the election (a string). Each value in the dictionary should contain the polls for one of the races.

```python
def build_frame(url):
    source = requests.get(url).text
    # Use StringIO because pd.DataFrame.from_csv requires .read() method
    s = StringIO(source)
    
    return pd.DataFrame.from_csv(s, index_col=None).convert_objects(
            convert_dates="coerce", convert_numeric=True)

# Makes a dictionary of pandas DataFrames keyed on election string.
polls = dict((election.split("/")[-1][:-4], build_frame(election)) for election in election_url)

#len(polls)
#polls['2014-kentucky-senate-mcconnell-vs-grimes']
del polls['2014-georgia-senate-runoff']
del polls['2014-new-hampshire-senate-bass-vs-shaheen']
polls.keys()
```

#### Problem 2(a)

Assume you have $M$ polls with sample sizes $n_1, \dots, n_M$. If the polls are independent, what is the average of the variances of each poll if the true proportion is $p$?

```python
rep_cand = list(candidates.R)
dem_cand = list(candidates.R)

race = []
race_state = []
mean_var = []
theo_std = []

for key, df in polls.iteritems():      
    if polls[key].columns[7] in rep_cand:
        p = (df[df.columns[7]] + df[df.columns[9]]/2)/100
    else:
        p = (df[df.columns[8]] + df[df.columns[9]]/2)/100   
        n = df[df.columns[4]]   
    df['variance'] = (p*(1-p)/n)
    df['std'] = np.sqrt(df['variance'])
    mean_var.append(np.mean(df['variance']))
    theo_std.append(np.mean(df['std']))
    race.append([key])
    racre_state.append(df['variance'])
    
headers = ['Race','Theo_Var','Theo_Std']
theoretical = pd.DataFrame([race, mean_var, theo_std], index=headers).T
theoretical.head()
print "Theoretical variaces are: ", mean_var
theoretical.head()
```

# Problem 3: Prediction and Posterior Probabilities

Perform a Bayesian analysis to predict the probability of Republicans winning in each state then provide a posterior distribution of the number of republicans in the senate.

#### Problem 3(a)

First, we define a Bayesian model for each race. The prior for the difference $d$ between republicans and democtrats will be $N(\mu,\tau^2)$. Say before seeing poll data you have idea who is going to win, what should $\mu$ be? How about $\tau$, should it be large or small? 

** Your answer here: **

Without seeing poll data but knowing the Rep will win, if μ = Rep - Dem, then μ is necessarily greater than zero and the τ should be smaller than μ/2.  This means there's 99.7% chance that μ will be greater than zero, meaning Rep gaining more votes than Dem.  

#### Problem 3(b)

What is the distribution of d conditioned on θ. What is the posterior distribution of θ|d?

**Hint**: Use normal approximation. 

**Your answer here:**

Distribution of d|θ is represented by N(θ,σ^2), centering at θ with the spread +- σ.  
σ = √ p(1-p)/N, 
where:
p is the estimated probability of vote, and N the poll sample size.

Posterior distribution of θ|d centers at Bμ+(1−B)d with the spread of 1/(1/σ^2 + 1/τ^2)
where: 
B =  σ^2 / (σ^2 + τ^2)
μ = historical lead
τ = SD of historical lead


#### Problem 3(c)

The prior represents what we think before hand. We do not know who is expected to win, so we assume $\mu=0$. For this problem estimate $\tau$ using the observed differences across states (Hint: $\tau$ represents the standard deviation of a typical difference). Compute the posterior mean for each state and plot it against original average. Is there much change? Why or why not? 

```python
rep_cand = list(candidates.R)
dem_cand = list(candidates.R)

election = []
margin = []
sig = []

for race, df in polls.iteritems():
    if polls[race].columns[7] in rep_cand:
        df['margin'] = (polls[race].ix[:, 7] - polls[race].ix[:, 8])/100
    else:
        df['margin'] = (polls[race].ix[:, 8] - polls[race].ix[:, 7])/100
    election.append([race])
    margin.append(np.mean(df['margin']))
    sig.append(np.std(df['margin']))
 
b_headers = ['Race', 'Margin','Obs_Std']

bayes = pd.DataFrame([election, margin, sig], index=b_headers).T

tau = np.std(bayes['Margin'])
sigma = bayes['Obs_Std']
mu = 0
bayes['B'] = sigma**2 / (sigma**2 + tau**2)

bayes.head()
```

```python
""" ANSWER IN CODE
τ = model by using σ   Rafael: "Yes we will use the same tau in all states"
Posterior mean is represented by θ|d
"""
B = bayes['B']
Y = bayes['Margin']
bayes['Posterior'] = B*mu + (1-B)*Y
bayes['Post_SD'] = (1/(1/(sigma**2) + 1/(tau**2)))**.5

bayes.head()
```

```python
plt.scatter(bayes['Margin'], bayes['Posterior'])
plt.xlim(-.6,0.6)
plt.ylim(-.6,0.6)
plt.plot([-.6,.6], [-.6,.6], linewidth=1, color='black')

plt.title("Senate Race 2014 Winning Margins")
plt.ylabel("Posterior Mean Margin")
plt.xlabel("Observed Mean Margin")
plt.gca().set_aspect('equal')
```

**Your answer here:**

The comparison between the posterior and the observed shows a slight decrease in the vote margins, along with a slight decrease in the SD.  Overall the posterior predicts a smaller winning margin but with a tighter confidence interval.

#### Problem 3(e)

Use the posterior distributions in a Monte Carlo simulation to generate election results. In each simulation compute the total number of seats the Republican control. Show a histogram of these results.

```python
post_mu = bayes.Posterior
post_SD = bayes.Post_SD
    
simulation = []

for j in range(5000):
    rep = 30  #number of seats the Republicans currently hold
    for i in range(len(post_mu)):  
        toss = np.random.normal(post_mu[i], post_SD[i])         
        if np.mean(toss) > 0:
            rep += 1

    simulation.append(rep)
    
plt.hist(simulation)
plt.title("Senate Race 2014: Predicted Republican Seats")
plt.ylabel("Number of Trials")
plt.xlabel("Number of Seats")
plt.show()
```

```python
def set_year(dateobj, year):
    d2 = dateobj.replace(year = year)
    return d2

wikitable['electionday'] = wikitable.apply(lambda r: set_year(r.values[1], r.values[0]), axis=1)
wikitable.head()
```

```python
#Grab all data
filedates = ["2010", "2012"]
pollstercsvs = [pd.read_csv("pollster_cleaned_%s.csv"%fname, parse_dates=['end.date']) 
                for fname in filedates]
#Paste them together
oldpolls = pd.concat(pollstercsvs)
oldpolls['end.date']

oldpolls = pd.merge(oldpolls, wikitable, how="inner", left_on="year", right_on="Year")

#Filter to senate races
oldpolls = oldpolls.ix[oldpolls.race.str.contains("Sen"),:]

#Compute number of days between poll's end date and election day.
oldpolls['day'] = oldpolls.apply(lambda x: (x['end.date']-x['electionday']).days, axis=1)

#Filter down to polls taken within a year of the election.
oldpolls = oldpolls.ix[(oldpolls.day <= 0) & (oldpolls.day >= -365),:]
```

```python
def grab_greater(df, focus, champ, chal):
    return np.where((df[chal] > df[focus]) & (df[champ] > df[focus]),
                    df[chal], df[focus])

oldpolls['real.dem2'] = grab_greater(oldpolls, 'real.dem', 'real.rep', 'real.oth')
oldpolls['real.rep2'] = grab_greater(oldpolls, 'real.rep', 'real.dem', 'real.oth')

oldpolls['vote.dem2'] = grab_greater(oldpolls, 'vote.dem', 'vote.rep', 'vote.oth')
oldpolls['vote.rep2'] = grab_greater(oldpolls, 'vote.rep', 'vote.dem', 'vote.oth')

oldpolls['diff'] = oldpolls['vote.rep2']-oldpolls['vote.dem2']
oldpolls['reald'] = oldpolls['real.rep2']-oldpolls['real.dem2']
oldpolls['bias'] = oldpolls['diff']-oldpolls['reald']

#print oldpolls.ix[oldpolls.bias.abs() >= 50,["vote.rep2", "vote.dem2", "real.rep2", "real.dem2"]]
oldpolls = oldpolls.ix[oldpolls.bias.abs() < 50,:]
```

```python
#For this analysis, consider only polls within a week of the election.
tmppolls = oldpolls.ix[oldpolls.day >= -60,]

#Compute number of polls taken by each pollster, in a format that lets you save this
#back to the original dataframe. Every poll corresponding to a given pollster will
#have the same value for 'npolls' -- `transform` creates these repeats for you.
tmppolls['npolls'] = tmppolls.groupby("pollster")['pollster'].transform(len)
#Only analyze pollsters with more than 2 polls.
tmppolls = tmppolls.ix[tmppolls.npolls >= 4,]

#Compute mean bias of each pollster. Save this back to the original dataframe.
tmppolls['mean_bias'] = tmppolls.groupby("pollster")['bias'].transform(np.mean)
#Make boxplot grouped by and sorted by mean bias.
tmppolls.boxplot("bias", by="mean_bias", rot=90)

#To get the right labels on the boxplot, recompute bias, but this time in a way
#that only returns one number per poll. `aggregate` only returns one number per
#pollster group.
pollster_bias = tmppolls.groupby("pollster")['bias'].aggregate(np.mean)
#Sort the pollster biases, then plot their indices to label the boxplot.
pollster_bias.sort()
plt.xticks(np.arange(len(pollster_bias)), pollster_bias.index, fontsize=10)
plt.ylim((-25,25))
plt.show()
```
